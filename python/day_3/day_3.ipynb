```python
allowed_numbers = []
i, j = 0, 0

array = lines
while i < array.shape[0]:
    while j < array.shape[1]:
        print("-" * 10)
        print(f"{i=},{j=}")
        e = array[i, j]
        print(f"{e=}")
        if e == ".":
            j += 1
            continue

        if e.isdigit():
            adjacent_to_symbol = False

            min_i = max(0, i - 1)
            max_i = min(i + 1, array.shape[0] - 1)
            min_j = max(0, j - 1)
            max_j = min(j + 1, array.shape[1] - 1)
            print(min_i, max_i, min_j, max_j)
            for a in range(min_i, max_i + 1):
                for b in range(min_j, max_j + 1):
                    temp_e = array[a, b]
                    print(f"{temp_e=}")
                    if (temp_e != ".") and not temp_e.isdigit():
                        adjacent_to_symbol = True
                        break

            if adjacent_to_symbol:
                print("adjacent!")

                acc = [e]
                # expand to the left :
                temp_j = j - 1
                while temp_j >= 0:
                    temp_e = array[i, temp_j]
                    if temp_e.isdigit():
                        acc.insert(0, temp_e)
                    else:
                        break
                    temp_j -= 1

                # expand to the right :
                temp_e = e
                temp_j = j + 1
                while temp_j <= array.shape[1]:
                    temp_e = array[i, temp_j]
                    if temp_e.isdigit():
                        acc.append(temp_e)
                    else:
                        break
                    temp_j += 1
                number = int("".join(acc))
                print(f"{number=}")
                allowed_numbers.append(number)

                j = temp_j
                continue

        j += 1
    j = 0
    i += 1
```

# Part 2

```python
def get_adjacent_star_coords(
    array: np.array, coords_x: [int,int], coords_y: [int,int]
) -> tuple[int, int] | None:
    min_i,max_i = coords_x
    min_j,max_j = coords_y
    for a in range(min_i, max_i + 1):
        for b in range(min_j, max_j + 1):
            temp_e = array[a, b]
            print(f"{temp_e=}")
            if temp_e=="*":
                return a,b
    return None
```

```python
part_numbers = defaultdict(list)
i, j = 0, 0

array = lines
while i < array.shape[0]:
    while j < array.shape[1]:
        print("-" * 10)
        print(f"{i=},{j=}")
        e = array[i, j]
        print(f"{e=}")
        if e == ".":
            j += 1
            continue

        if e.isdigit():
            adjacent_to_star = False

            min_i = max(0, i - 1)
            max_i = min(i + 1, array.shape[0] - 1)
            min_j = max(0, j - 1)
            max_j = min(j + 1, array.shape[1] - 1)
            print(min_i, max_i, min_j, max_j)
            star_coords = get_adjacent_star_coords(
                array, [min_i,max_i],[min_j,max_j]
            )
            

            if star_coords is not None:
                
                print(f"adjacent to {star_coords}")

                acc = [e]
                # expand to the left :
                temp_j = j - 1
                while temp_j >= 0:
                    temp_e = array[i, temp_j]
                    if temp_e.isdigit():
                        acc.insert(0, temp_e)
                    else:
                        break
                    temp_j -= 1

                # expand to the right :
                temp_e = e
                temp_j = j + 1
                while temp_j <= array.shape[1]:
                    temp_e = array[i, temp_j]
                    if temp_e.isdigit():
                        acc.append(temp_e)
                    else:
                        break
                    temp_j += 1
                number = int("".join(acc))
                print(f"{number=}")
                part_numbers[star_coords].append(number)

                j = temp_j
                continue

        j += 1
    j = 0
    i += 1
```

```python
part_numbers
```

```python
total = 0

for parts in part_numbers.values():
    if len(parts)>1:
        total += np.prod(parts)
```
