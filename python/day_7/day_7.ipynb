# Part 1

```python
cards_values = dict(zip("AKQJT98765432", range(14, 1, -1)))
```

```python
def get_power(x: str) -> int:
    counts = Counter(x).most_common(5)
    power = 0
    if counts[0][1] == 5:  # Five of a kind
        power = 6
    elif counts[0][1] == 4:
        power = 5
    elif (counts[0][1] == 3) and (counts[1][1] == 2):
        power = 4
    elif (counts[0][1] == 3) and (counts[1][1] == 1):
        power = 3
    elif (counts[0][1] == 2) and counts[1][1] == 2:
        power = 2
    elif counts[0][1] == 2:
        power = 1
    return power
```

```python
def get_result(input_list: list[list[str, str]]) -> int:
    sorted_ = sorted(
        sorted(input_list, key=lambda x: tuple(cards_values[e] for e in tuple(x[0]))),
        key=lambda x: get_power(x[0]),
    )
    result = 0
    for rank, val in enumerate(sorted_):
        result += int(val[1]) * (rank + 1)
    return sorted_,result
```

```python
get_result(input_)
```

```python
cards_values = dict(zip("AKQT98765432J", range(14, 1, -1)))
```

```python
def get_power(x: str) -> int:
    counter = Counter(x)
    counts = Counter(x).most_common(5)
    if "J" in counter and len(counter)>1:
        del counter['J']
        most_common = counter.most_common(1)
        x= x.replace("J",most_common[0][0])
        counter = Counter(x)
        counts = Counter(x).most_common(5)
        
    power = 0
    if counts[0][1] == 5:  # Five of a kind
        power = 6
    elif counts[0][1] == 4:
        power = 5
    elif (counts[0][1] == 3) and (counts[1][1] == 2):
        power = 4
    elif (counts[0][1] == 3) and (counts[1][1] == 1):
        power = 3
    elif (counts[0][1] == 2) and counts[1][1] == 2:
        power = 2
    elif counts[0][1] == 2:
        power = 1
    return power
```

```python
def get_result(input_list: list[list[str, str]]) -> int:
    sorted_ = sorted(
        sorted(input_list, key=lambda x: tuple(cards_values[e] for e in tuple(x[0]))),
        key=lambda x: get_power(x[0]),
    )
    result = 0
    for rank, val in enumerate(sorted_):
        result += int(val[1]) * (rank + 1)
    return sorted_,result
```
