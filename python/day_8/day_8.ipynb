# Part 1

```python
def create_input_dict(lines: list[str]) -> tuple[str, dict]:
    res = {}

    first_key = None
    for line in lines:
        key, directions = line.split(" = ")

        if first_key is None:
            first_key = key

        res[key] = directions[1:-1].split(", ")

    return first_key, res
```

```python
def create_input_dict_2(lines: list[str]) -> tuple[list[str], dict]:
    res = {}

    first_keys = []
    for line in lines:
        key, directions = line.split(" = ")

        if key[-1]=='A':
            first_keys.append(key) 

        res[key] = directions[1:-1].split(", ")

    return first_keys, res
```

```python
first_keys, map = create_input_dict_2(lines[2:])
```

```python
def find_ends(key,map,limit=4):
    key = key
    i = 0
    steps = 0
    ends = []
    while len(ends)<limit:
        if i == (len(order)):
            i = 0
    
        dir = order[i]
        #print(key)
        if dir == "L":
            key = map[key][0]
        else:
            key = map[key][1]
        
        
        i += 1
        steps += 1
        if key[-1]=='Z':
            ends.append((key,steps))
    return ends
```
